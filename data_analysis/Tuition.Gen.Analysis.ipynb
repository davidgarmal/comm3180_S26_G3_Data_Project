## Tuition By State Ranked

## Changes from 21-22 to 22-23 for each school type

```python
states = states.copy()

states["pub4_change"] = states["pub4_instate_tf_2223"] - states["pub4_instate_tf_2122"]
states["priv4_change"] = states["priv4_tf_2223"] - states["priv4_tf_2122"]
states["pub2_change"] = states["pub2_instate_tf_2223"] - states["pub2_instate_tf_2122"]

print("Average tuition increase from 2021-22 to 2022-23:")
print("Public 4-year in-state: $", round(states["pub4_change"].mean(), 2))
print("Private 4-year:         $", round(states["priv4_change"].mean(), 2))
print("Public 2-year in-state: $", round(states["pub2_change"].mean(), 2))

print("\nPublic 4-year change by state:")
pub4_yoy = states[["state", "pub4_instate_tf_2122", "pub4_instate_tf_2223", "pub4_change"]].sort_values("pub4_change", ascending=False)
print(pub4_yoy.to_string(index=False))

print("\nPrivate 4-year change by state:")
priv4_yoy = states[["state", "priv4_tf_2122", "priv4_tf_2223", "priv4_change"]].sort_values("priv4_change", ascending=False)
print(priv4_yoy.to_string(index=False))

print("\nPublic 2-year change by state:")
pub2_yoy = states[["state", "pub2_instate_tf_2122", "pub2_instate_tf_2223", "pub2_change"]].sort_values("pub2_change", ascending=False)
print(pub2_yoy.to_string(index=False))
```
