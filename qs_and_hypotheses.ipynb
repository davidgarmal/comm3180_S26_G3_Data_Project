# Questions and Hypotheses: 

**Diversity, Place, and Economic Mobility: Analyzing the Relationship Between School and Community Demographics and Post-Graduate Outcomes**

This project investigates how the racial and demographic composition of both universities and their surrounding communities shapes students' economic trajectories after graduation. Using three complementary datasets, the U.S. Department of Education's College Scorecard and U.S. Census tract-level data [and Jenna's dataset here], we examine whether institutional and neighborhood diversity predict post-graduate earnings, and whether community demographics can explain variation in the diversity of student bodies across colleges.

**Hypothesis 1 (College Scorecard):** Students who attend more racially and demographically diverse universities will demonstrate higher median earnings after graduation, reflecting the broad economic and social benefits of diverse educational environments.[^1] Using the College Scorecard, we will examine how institutional racial composition relates to post-graduate earning outcomes. Our primary research question is: *Do the racial and demographic characteristics of a university predict the median earnings of its graduates?* We assume that greater racial homogeneity will have adverse outcomes for students at primarily white institutions (though not necessarily at HBCUs, for example). We will also investigate whether this relationship is confounded by systemic factors, such as family income distribution, and completion rates at the university overall, that affect earning potential independently of diversity itself.

**Hypothesis 2 (U.S. Census):** Colleges located in more racially diverse census tracts will have more diverse student bodies, reflecting the demographic composition of the communities in which they are embedded. Using Census tract-level data on racial and ethnic composition (including percent Black, AAPI, AIAN, Hispanic, and White residents), we will ask: *Does the diversity of a college's surrounding community predict the diversity of its student body?* We expect a positive relationship, but anticipate that community income levels and institutional cost of attendance will act as meaningful confounders — potentially dampening diversity even in demographically heterogeneous areas. As a secondary line of inquiry connecting back to Hypothesis 1, we will explore whether schools in more diverse communities are also associated with lower post-graduate earnings, and whether this reflects local economic conditions and post-graduation geographic mobility rather than institutional quality. The feasibility of the mobility analysis will depend on data availability within the Census.
